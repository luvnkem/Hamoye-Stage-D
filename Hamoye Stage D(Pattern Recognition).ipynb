## Data StoryTelling in Pattern Recognition 
## Africa Food Production and Supply(2004-2013)

### Importing the Relevant Libraries

```python
import pandas as pd
```

```python
# Importing the food production dataset

food_prod = pd.read_csv('Africa Food Production (2004 - 2013).csv', delimiter=',')
food_prod.head()
```

```python
# Also importing the food supply dataset
# Opening and viewing the first 5 rows of the food supply dataset

food_sup = pd.read_csv('Africa Food Supply (2004 - 2013).csv')
food_sup.head()
```

## Question 2
Which three countries had the lowest food supply between 2004 and 2006 respectively(inclusive)?  

```python
# Grouping the food supply dataset by country and year to get the total value of items supplied

food_sup_sum= pd.DataFrame(food_sup.groupby(['Country', 'Year'], as_index = False).sum())
food_sup_sum
```

```python
# Sorting by total values supplied in descending order
food_sup_sorted = food_sup_sum.sort_values(by='Value', ascending=False)

food_sup_sorted.tail(32)
```

### The 3 countries with the lowest food supply between 2004 and 2006 are Zambia, Central African Republic and Ethopia

## Question 5
What three countries have produced the highest total quantity of rice from 2004 to 2013?

```python
# Extracting the data for rice production in African countries from 2004 to 2013
rice_data = food_prod.query("Item == 'Rice (Milled Equivalent)'")
 
rice_data
```

```python
# Finding the sum of values per country per year
rice_total = pd.DataFrame(rice_data.groupby(['Country', 'Year'], as_index = False)['Value'].sum())
rice_total.head()
```

```python
# Sorting the data by total amount of Rice produced in descending order
rice_sorted = rice_total.sort_values(by='Value', ascending=False)

rice_sorted.head(32)
```

### The 3 countries that produced the highest number of Rice from 2004 t0 2013 are Egypt, Nigeria and Madagascar

## Question 4
Which country produced the highest amount of oats in 2010?

```python
# Extracting the entire data for Oats production in African countries from 2004 to 2013
Oat_data = food_prod.query("Item == 'Oats'")
 
Oat_data
```

```python
# Finding the sum of values per country per year
oat_total = pd.DataFrame(Oat_data.groupby(['Country', 'Year'], as_index = False)['Value'].sum())
oat_total.head()
```

```python
# Sorting the data by total amount of Oat produced in descending order
oat_sorted = oat_total.sort_values(by='Value', ascending=False)

oat_sorted.head(50)
```

### The country that produced the highest amount of Oats in 2010 is Algeria
